### about the Arab World

```python
df_pr = df[df["Country Name"]=='Arab World']
```

```python
df_pr.plot(kind = 'line',
           x = 'Year',
           y = 'Value',
           figsize = (30,15),
           legend = True,
           grid = True,
          xlabel = 'Year',
          ylabel = 'GDP growth',
          title = "GDP ANALYSIS")
```

```python
# 1968 - 2.576
# 1969 - 2.843
# 1970 - 3.138
```

```python
# factor by which gdp is increasing 
```

```python
round(((2.843-2.576)/2.576)*100,2)    #increasing by 10 percent
```

```python
data = df_pr.values
gdp_change = [0]
for i in range(1,len(data)):
   gdp_change.append(round((data[i][3]-data[i-1][3])/data[i-1][3]*100,2))
    
```

```python
len(gdp_change)
```

```python
df_pr.assign(GDP = gdp_change)
```
